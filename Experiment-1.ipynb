```python
# importing numpy package
import numpy as np

# importing matplotlib package
import matplotlib.pyplot as plt

# importing mplot3d from mpl_toolkits
from mpl_toolkits import mplot3d

#importing json
import json
```

```python
# this sets up the Matplotlib interactive windows
%matplotlib widget
```

```python
# creating an empty canvas
fig = plt.figure()
 
# defining the axes with the projection
# as 3D so as to plot 3D graphs
ax = plt.axes(projection="3d")
```

```python
# load data from file
with open("data/sample-1.json", "r") as datafile:
  points = json.load(datafile)
```

```python
# points x,y,z
x=points["x"]
y=points["y"]
zf=points["zf"]
zr=points["zr"]
```

```python
# plotting a 3D line graph with X-coordinate,
# Y-coordinate and Z-coordinate respectively
ax.plot3D(x, y, zf, 'blue')
ax.plot3D(x, y, zr, 'red')
```

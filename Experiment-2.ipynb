```python
# importing numpy package
import numpy as np

# importing matplotlib package
import matplotlib.pyplot as plt

#importing json
import json
```

```python
# this sets up the Matplotlib interactive windows
%matplotlib widget
```

```python
# creating an empty canvas
fig = plt.figure()

# 2D axes
ax = plt.axes()
ax.set_aspect('equal','box')
```

```python
# load data from file
with open("data/sample-1.json", "r") as datafile:
  points = json.load(datafile)

with open("data/sample-2.json", "r") as datafile2:
  points2 = json.load(datafile2)
```

```python
# points x,y
x=points["x"]
y=points["y"]

x2=points2["x"]
y2=points2["y"]
```

```python
ax.plot(x, y, '-b')
ax.plot(x2, y2, '-r')
ax.plot(0,0, '+k')
```
